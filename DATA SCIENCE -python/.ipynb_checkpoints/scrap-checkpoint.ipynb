#### SETUP

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud,STOPWORDS
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer as sia
```

#### DATA EXTRACTION 

```python
head = []
rating = []
names  = []
country = []
date = []
comment = []
aircraft = []
other_info = []



for i in range(1,374):
    #print(f"Scraping page {i}")
    url = f"https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    page = requests.get(url, headers=headers, timeout=100)
   

    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find_all(itemprop="review")

    
    for i in results:

        header = i.find('h2', class_ = 'text_header')
        head.append(header.text)
        rate = i.find(itemprop = 'ratingValue')
        if rate:
            rating.append(rate.extract().text)
        else:
            rating.append('None')
        name = i.find(itemprop = 'name')
        names.append(name.text)
        country_get = i.find('h3' , class_ = 'text_sub_header userStatusWrapper')
        substring = '('
        if substring in country_get.text:
            clean_step_one= country_get.text
            clean_step_two = clean_step_one.split(')')
            clean_step_three =str(clean_step_two[0])
            clean_step_four= clean_step_three.split('(') 
            country.append(clean_step_four[1])
        else:
            country.append('None')
        date_get = i.find('time', itemprop = 'datePublished')
        date.append(date_get.text)
        comment_get = i.find(class_ = 'text_content')
        comment.append(comment_get.text)
        aircraft_get = i.find(class_ = 'review-rating-header aircraft')
        #aircraft.append(aircraft_get.text)


        rheaders = i.find_all('td', class_='review-rating-header')
        rvalues = i.find_all('td', class_='review-value')

        extracted_data = {}

        for header, value in zip(rheaders, rvalues):
            header_text = header.text.strip().lower()
            value_text = value.text.strip()

            # Check if the header matches any specified header
            for specified_header in ['aircraft', 'type of traveller', 'seat type', 'route', 'date flown']:
                if specified_header in header_text:
                    extracted_data[specified_header.replace(' ', '_')] = value_text
                    break
            else:
                # If no specified header is found, set it to None
                extracted_data[header_text.replace(' ', '_')] = None

        other_info.append(extracted_data)
```

```python
list_to_dict = {'name':names, 'country':country, 'date':date, 'header':head, 'review':comment, 'rating':rating}
```

```python
data = pd.DataFrame(list_to_dict)
```

```python
data2 = pd.DataFrame(other_info)
```

#### MERGE DATASETS EXTRACTED

#### SENTIMENT ANALYSIS

```python
df.review
```

```python
score = [] 

for text in df.review:
    pol_score = sia().polarity_scores(text)
    score.append(pol_score)
```

```python
df['sentiment'] = pd.DataFrame(score)['compound']
```

```python
df.sentiment = round(df.sentiment)
```
